## 2️⃣ Preprocesamiento de los Datos

### 🔄 Transformar las variables que representan montos monetarios, de texto a numéricas.

```python
# Antes de realizar algún cambio realizo una copia
df_original = df.copy()

# Transformo las variables de montos, que son texto, a variables numéricas enteras
# Defino las valriables monetarias a modificar
monetarias = [
    'Income',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds',
    'MntTotal'
]

# Reemplazo de valores

df[monetarias] = (
    df[monetarias]
    .replace({'\$': '', ',': ''}, regex=True)
    .apply(pd.to_numeric, errors='coerce')
    .astype('Int64')
)

# Validación
df[monetarias].info()
```

## 🛠️ Reconstrucción de variables categóricas

```python
# Reconstrucción de variable Marital_Status a partir de las variables binarias
marital_cols = [
    'marital_Single',
    'marital_Married',
    'marital_Together',
    'marital_Divorced',
    'marital_Widow'
]

# Validar que existe un sólo estado por cliente
df[marital_cols].sum(axis=1).value_counts()

#Crear la nueva variable
df['Marital_Status'] = (
    df[marital_cols]
    .idxmax(axis=1)
    .str.replace("marital_", "")
)

# Validación
single1 = (df['Marital_Status'] == 'Single').sum()
single2 = (df['marital_Single'] == 1).sum()
married1 = (df['Marital_Status'] == 'Married').sum()
married2 = (df['marital_Married'] == 1).sum()
together1 = (df['Marital_Status'] == 'Together').sum()
together2 = (df['marital_Together'] == 1).sum()
divorced1 = (df['Marital_Status'] == 'Divorced').sum()
divorced2 = (df['marital_Divorced'] == 1).sum()
widow1 = (df['Marital_Status'] == 'Widow').sum()
widow2 = (df['marital_Widow'] == 1).sum()

print('Soltero', single1, single2)
print('Casado', married1, married2)
print('Unión libre', together1, together2)
print('Divorciado', divorced1, divorced2)
print('Viudo', widow1, widow2)

total = single1 + married1 + together1 + divorced1 + widow1
print('Total', total)
```

```python
# Reconstrucción de variable Education a partir de las variables binarias
education_cols = [
    'education_Basic',
    'education_Graduation',
    'education_Master',
    'education_PhD',
    'education_2n Cycle'
]

# Validar que existe un sólo estado por cliente
df[education_cols].sum(axis=1).value_counts()

#Crear la nueva variable
df['Education'] = (
    df[education_cols]
    .idxmax(axis=1)
    .str.replace("education_", "")
)

# Validación
basic1 = (df['Education'] == 'Basic').sum()
basic2 = (df['education_Basic'] == 1).sum()
graduation1 = (df['Education'] == 'Graduation').sum()
graduation2 = (df['education_Graduation'] == 1).sum()
master1 = (df['Education'] == 'Master').sum()
master2 = (df['education_Master'] == 1).sum()
phd1 = (df['Education'] == 'PhD').sum()
phd2 = (df['education_PhD'] == 1).sum()
cycle1 = (df['Education'] == '2n Cycle').sum()
cycle2 = (df['education_2n Cycle'] == 1).sum()

print('Básica', basic1, basic2)
print('Graduación', graduation1, graduation2)
print('Maestría', master1, master2)
print('PhD', phd1, phd2)
print('2do ciclo', cycle1, cycle2)

total = basic1 + graduation1 + master1 + phd1 + cycle1
print('Total', total)
```
